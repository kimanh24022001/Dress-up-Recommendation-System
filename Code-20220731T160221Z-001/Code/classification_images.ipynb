### **IMPORT THƯ VIỆN**

```python
import cv2 as cv
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pickle
```

### **TRAIN TEST SPLIT**

```python
import os
dir_dataset= '/gdrive/MyDrive/Đồ án Nhập môn Thị giác máy tính (CS231.M22.KHCL)/Dataset'
```

```python

dir_images_train = os.path.join(dir_dataset,'Train')
dir_images_test = os.path.join(dir_dataset,'Test')
```

### **HÀM TÍNH HISTOGRAM**

```python
def TinhHist(pathfilename):
  img = cv.imread(pathfilename,0)
  hist = cv.calcHist([img],[0],None,[256],[0,256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size
  return hist                   
```

### **ĐỌC DATA**

```python
def ReadData(path_to_files):
    labels = []
    features = []
    for file_name in os.listdir(path_to_files):
        type=['female','male','unisex']
        for j in type:
            path_1=os.path.join(path_to_files, j)
            if j=='female':
                list=['top','bottom','bag&backpack','dress','shoes','hat']
                for i in list:
                    path_2=os.path.join(path_1, i)
                    for file_name in os.listdir(path_2):
                        feature=TinhHist(os.path.join(path_2, file_name)).reshape(-1)
                        features.append(feature)
                        label='female_'+i
                        labels.append(label)
            if j=='male':
                list=['top','bottom','bag&backpack','shoes','hat']
                for i in list:
                    path_2=os.path.join(path_1, i)
                    for file_name in os.listdir(path_2):
                        feature=TinhHist(os.path.join(path_2, file_name)).reshape(-1)
                        features.append(feature)
                        label='male_'+i
                        labels.append(label)
            if j=='unisex':
                list=['bottom','bag&backpack','shoes','hat']
                for i in list:
                    path_2=os.path.join(path_1, i)
                    for file_name in os.listdir(path_2):
                        feature=TinhHist(os.path.join(path_2, file_name)).reshape(-1)
                        features.append(feature)
                        label='unisex_'+i
                        labels.append(label)               
    return np.array(features), np.array(labels)
```

### **ĐỌC DATA TẬP TRAIN,TEST**

```python
train_features, train_labels = ReadData(dir_images_train)

test_features, test_labels = ReadData(dir_images_test)
```

```python
train_features 
```

### **DÙNG MODEL KNN PHÂN LOẠI NHIỀU LỚP**

```python
model = KNeighborsClassifier(n_neighbors=5)

model.fit(train_features, train_labels)
```

### **ĐỘ CHÍNH XÁC MODEL, DỰ ĐOÁN**

```python
acc = model.score(test_features, test_labels)
pre=model.predict(test_features)

print("Độ chính xác khi K=5: ",acc,pre,test_labels)
```
